```python
import pandas as pd
```

```python
data = pd.read_csv("Data\housing.csv")
```

```python
data
```

```python
#features
X = data.drop('median_house_value',axis=1)
#output
y = data.iloc[:,8:9]
```

```python
# filling null values in total_bedrooms field with the mean
X['total_bedrooms'] = X['total_bedrooms'].fillna(X['total_bedrooms'].mean())
```

```python
X['total_bedrooms'] = X['total_bedrooms'].fillna(X['total_bedrooms'].median())
```

```python
#encoding categorical data
X = pd.get_dummies(data=X)
```

```python
from sklearn.model_selection import train_test_split


#split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
#standarization numerical values since it's too big


from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_test
```

```python
from sklearn import metrics
metrics.r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred )
```
